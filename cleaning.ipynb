```python
import pandas as pd
df = pd.read_csv("California_Fires.csv",skiprows = 0)
df.head(5)
```

```python
df.columns
```

```python
dfclean = df[['incident_administrative_unit','incident_county', 'incident_acres_burned',
       'incident_cooperating_agencies', 'incident_dateonly_extinguished', 'incident_dateonly_created']]
```

```python
dfclean
```

```python
dfclean['fire_started'] = pd.to_datetime(dfclean['incident_dateonly_created'], format='%m/%d/%y')
```

```python
dfclean['fire_extinguished'] = pd.to_datetime(dfclean['incident_dateonly_extinguished'], format='%m/%d/%y')
```

```python
dfclean.column
```

```python
dfclean['days_of_fire'] = ( dfclean['fire_extinguished'] - dfclean['fire_started']).dt.days
```

```python
dfclean.head()
```

```python
df_final = dfclean[['incident_county','incident_acres_burned', 'incident_cooperating_agencies',
       'fire_started', 'fire_extinguished', 'days_of_fire']]
```

```python
df_final
```

```python
df_final.to_csv('calFire_cleaned.csv',index=False)
```
